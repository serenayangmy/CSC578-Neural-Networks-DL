```````````python
#part 3 C.a
import numpy as np
data = iris_data
np.random.seed(12793)#random the samples with seed
np.random.shuffle(data)

train_data = data[0:int(len(data)*0.7)] #first 70% randomed samples
test_data = data[int(len(data)*0.7):]#second randomed 30% samples
```````````

```````````python
#part 3 C.b
# Create a new network
net4 = network.Network([4,20,3])
```````````

```````````python
#part 3 C.c
results = net4.SGD(train_data,50,5,0.1,test_data)

train = results[0]
test = results[1]

num_epoch, mse_train, ce_train, ll_train, mse_test, ce_test, ll_test = [], [], [], [], [], [],[]

for i in range (50):
    num_epoch.append(i)
    
    #Train 
    mse_train.append(train[i][2])
    ce_train.append(train[i][3])
    ll_train.append(train[i][4])
    
    #Test
    mse_test.append(test[i][2])
    ce_test.append(test[i][3])
    ll_test.append(test[i][4])
    
```````````

```````````python
#part 3 D.a
import matplotlib.pyplot as plt
import seaborn as sns
```````````

```````````python
plt.ylim(0.0, 2)
plt.plot(num_epoch, mse_train, color='red', label="MSE")
plt.plot(num_epoch, ce_train, color='green', label="CrossEntropy")
plt.plot(num_epoch, ll_train, color='blue', label="LogLikelihood")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Error by Cost Function (training)')
plt.legend()
plt.grid(linestyle='-', linewidth='0.2', color='gray')
```````````

```````````python
#part 3 D.b
plt.ylim(0.0, 0.4)
plt.plot(num_epoch, mse_train, color='red', label="Train")
plt.plot(num_epoch, mse_test, color='orange', label="Test")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Train vs Test: MSE')
plt.legend()
```````````

```````````python
plt.ylim(0.0, 2)
plt.plot(num_epoch, ce_train, color='green', label="Train")
plt.plot(num_epoch, ce_test, color='orange', label="Test")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Train vs Test: CrossEntropy')
plt.legend()
```````````

```````````python
plt.ylim(0.0, 1)
plt.plot(num_epoch, ll_train, color='blue', label="Train")
plt.plot(num_epoch, ll_test, color='orange', label="Test")
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Train vs Test: LogLikelihood')
plt.legend()
```````````
